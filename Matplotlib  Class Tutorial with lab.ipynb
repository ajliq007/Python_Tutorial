 ### Line Plots (Graphs)
   
A line plot is a type of chart which displays information as a series of data points called 'markers' connected by straight line segments. Line graphs are used to track changes over short and long periods of time. Line graphs can also be used to compare data sets over the same period of time for more than one group (a series).
- X-axis requires quantitative variable
- Y-axis values must be numbers and can be set to be interpreted (formatted) as decimal, percent, and currency.
- Variables have contiguous values
- Familiar/conventional ordering among ordinals

<img src="images/line_plot_example1.jpg" width="500">


## Data set for our line plot
For the data set for our line plots we will use the **numpy** library function ``arrange`` to give us generated data for plotting. Additionally, we will include **math**, which is the Python Math Library and provides us access to some common math functions and constants.

````python
import numpy as np
import math
# the x axis: ndarray object of angles between 0 and 2π
x = np.arange(0, math.pi*2, 0.05)
# the y axis: sine values of angles in x to be displayed on y axis
y = np.sin(x)
````

## Single Series line plot
First, we create a simple line plot of our data using **pyplot**.

````python
import matplotlib.pyplot as plt

plt.plot(x,y)
plt.show()
````

````python
plt.plot(x,y,'g')
plt.show()
````

Then we will add some labels and markers to make the plot presentable.

````python
#You can set the plot title, and labels for x and y axes.
plt.xlabel("angle")
plt.ylabel("sine")
plt.title('sine wave')
plt.plot(x,y)
plt.show()
````

## Multiple Series line plot
In this case we will add in the ``cosine`` as our second dataset so that we can demonstrate how to plot a line series. NOTE: we will use the same x ndarray, as we want to keep the same X-Axis values.
- use the same x values
- add the second y array (cosine)
- add the labels
- add a legend

````python
# the y axis for Cosine: cosine values of angles in x to be displayed on y axis
y_cos = np.cos(x)
#You can set the plot title, and labels for x and y axes.
plt.xlabel("angle")
plt.ylabel("sine/cosine values")
plt.title('sine/cosine wave')
plt.plot(x,y,x,y_cos)
plt.legend(['sin(x)', 'cos(x)'])      # legend entries as seperate strings in a list
plt.show()
````

 ### Bar Plots 
 
A barplot (or barchart) is one of the most common types of graphic. It shows the relationship between a numeric and a categoric variable. Each entity of the categoric variable is represented as a bar. The size of the bar represents its numeric value.   
- Bar graphs are used to compare data between different groups or to track changes over time (similar to a line chart). 
- Bar graphs can display multiple groups in the same bar in a stacked visual showing the composition of the bar chart along the Y-axis.

<img src="images/bar_plot_example1.jpg" width="500">

Matplotlib provides the **bar()** function for creating bar plots. 

The **bar()** function parameters to be used with axes object is as follows −

``bar(x, height, width, bottom, align)``


## Data set for a bar plot
In this example we will keep the data simple. We will use an example of number of students enrolled for various courses at a college. 


````python
# define the data set
courses = ['Data Science', 'Python', 'R', 'Statistics', 'Machine Learning']
students = [30, 45, 8, 25, 38]
````

## The standard bar plot
Recall what you have learned:
- Figures
- Axes
- Plots

The following code will create a bar chart using student enrollment data.


````python
# use pyplot for our plotting constructs
import matplotlib.pyplot as plt
# create a figure to "draw" on
fig = plt.figure()
# add our subplot using rectangle coordinates [left, bottom, width, height] 
ax = fig.add_axes([0,0,1,1])
# add the bar chart to the subplot
ax.bar(courses, students)
# add our labels and a title
plt.xlabel("Courses")
plt.ylabel("Student Enrollment")
plt.title('Student Enrollment by Course')
# display the plot
plt.show()
````

# List of named colors
https://matplotlib.org/3.1.0/gallery/color/named_colors.html

````python
import matplotlib.pyplot as plt
import matplotlib.colors as mcolors


def plot_colortable(colors, title, sort_colors=True, emptycols=0):

    cell_width = 212
    cell_height = 22
    swatch_width = 48
    margin = 12
    topmargin = 40

    # Sort colors by hue, saturation, value and name.
    if sort_colors is True:
        by_hsv = sorted((tuple(mcolors.rgb_to_hsv(mcolors.to_rgb(color))),
                         name)
                        for name, color in colors.items())
        names = [name for hsv, name in by_hsv]
    else:
        names = list(colors)

    n = len(names)
    ncols = 4 - emptycols
    nrows = n // ncols + int(n % ncols > 0)

    width = cell_width * 4 + 2 * margin
    height = cell_height * nrows + margin + topmargin
    dpi = 72

    fig, ax = plt.subplots(figsize=(width / dpi, height / dpi), dpi=dpi)
    fig.subplots_adjust(margin/width, margin/height,
                        (width-margin)/width, (height-topmargin)/height)
    ax.set_xlim(0, cell_width * 4)
    ax.set_ylim(cell_height * (nrows-0.5), -cell_height/2.)
    ax.yaxis.set_visible(False)
    ax.xaxis.set_visible(False)
    ax.set_axis_off()
    ax.set_title(title, fontsize=24, loc="left", pad=10)

    for i, name in enumerate(names):
        row = i % nrows
        col = i // nrows
        y = row * cell_height

        swatch_start_x = cell_width * col
        swatch_end_x = cell_width * col + swatch_width
        text_pos_x = cell_width * col + swatch_width + 7

        ax.text(text_pos_x, y, name, fontsize=14,
                horizontalalignment='left',
                verticalalignment='center')

        ax.hlines(y, swatch_start_x, swatch_end_x,
                  color=colors[name], linewidth=18)

    return fig

plot_colortable(mcolors.BASE_COLORS, "Base Colors",
                sort_colors=False, emptycols=1)
plot_colortable(mcolors.TABLEAU_COLORS, "Tableau Palette",
                sort_colors=False, emptycols=2)

#sphinx_gallery_thumbnail_number = 3
plot_colortable(mcolors.CSS4_COLORS, "CSS Colors")

# Optionally plot the XKCD colors (Caution: will produce large figure)
#xkcd_fig = plot_colortable(mcolors.XKCD_COLORS, "XKCD Colors")
#xkcd_fig.savefig("XKCD_Colors.png")

plt.show()

````

## The multi-series bar plot
In this example, we will keep the X axis data the same (i.e, the courses). We will add in a new enrollment facet of "campus" which will give us three total data sets (or series), those being: "Main campus", "North campus", and "Virtual campus".

To augment the plot, we will:
- Add color to distinguish the data sets (bar colors)
- The labels will remain the same

````python
# import numpy so we have access to the arrange() function
import numpy as np
#creat our data - Main, North, Virtual campus (array 1, 2, 3 respectively)
students_by_campus = [[30, 45, 8, 25, 38],[15, 20, 19, 23, 10],[45, 35, 40, 33, 39]]
# create a figure to "draw" on
fig = plt.figure()
# add our subplot using rectangle coordinates [left, bottom, width, height] 
ax = fig.add_axes([0,0,1,1.5])
# arrange the position of the bar charts by category by using an offset
X = np.arange(5)
# create the bars
ax.bar(X + 0.00, students_by_campus[0], color = 'b', width = 0.25)
ax.bar(X + 0.25, students_by_campus[1], color = 'g', width = 0.25)
ax.bar(X + 0.50, students_by_campus[2], color = 'r', width = 0.25)
# add our labels and a title
plt.xlabel("Courses")
plt.ylabel("Student Enrollment")
plt.title('Student Enrollment by Course by Campus')
# set the x ticks and labels
ax.set_xticks(X+.25)
ax.set_xticklabels(courses)
# create the legend
ax.legend(labels=['Main', 'North', 'Virtual'])
# display the plot
plt.show()
````

 ### Definition of a Scatter Plot 
   
- A Scatter Plot is used to determine relationships between two different data dimensions. 
- The x-axis is used to measure one dimension (or variable) and the y-axis is used to measure the other. If both variables increase at the same time, they have a positive relationship. If one variable decreases while the other increases, they have a negative relationship. Sometimes the variables don't follow any pattern and have no relationship.

<img src="images/scatter_plot_example.png" width="500">

Matplotlib provides the **scatter()** function for creating scatter plots. 

The **scatter()** function parameters to be used with axes object are as follows (not a complete list) −

``scatter(x_coords, y_coords, shape, colors, alpha)``

## Data for a scatter plot

For the first data set, we will use randomly generated data. As you recall, **numpy** can provide these types of functions. In this case we will use the ``rand()`` function to randomly generate the data that we will use for the scatter plot. 

We will be using 500 points in our data set (an ndarray).


````python
# include numpy for the rand() function
import numpy as np
# Create data
N = 500 # we want 500 points
x = np.random.rand(N)
y = np.random.rand(N)
````

For the second data set, we will also use the ``rand()`` function, but we will create 3 groups of 100 points 

````python
# Create data for the three groups (of ndarray tuples) of data we will be plotting
N = 100 # we want 100 points per dataset
# to create separation in the random sets, add in X and Y offset values
group1 = (0.6 + 0.6 * np.random.rand(N), np.random.rand(N))
group2 = (0.4+0.3 * np.random.rand(N), 0.5*np.random.rand(N))
group3 = (0.3*np.random.rand(N),0.3*np.random.rand(N))
````

## Creating the scatter plot
We will use the ``scatter`` function with the data that we created to create our scatter plot. 

Notice a few things:
- first, we did not create a figure or add axes
- this means that we are using the defaults
- second, we don't have to add the scatter plot to the axes (or subplot)


````python
# import Pyplot
import matplotlib.pyplot as plt

# Create the scatter plot by using the scatter function
plt.scatter(x, y, c='g', alpha=1)
# Add in the title
plt.title('Scatter plot using MatplotLib')
# Add in the labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
````

This time let's try it with the creation of the Figure and the Axes.

The result is the same, but it enforces the concepts of how matplotlib "draws."

````python
# define the colors and shape (area) parameters
colors = (0,0,0)
area = np.pi*3

# create a figure to "draw" on
fig = plt.figure()
# add our subplot using rectangle coordinates [left, bottom, width, height] 
ax = fig.add_axes([0,0,.8,.8])
# add the scatter plot to the subplot
plt.scatter(x, y, c='g', alpha=1)

# Add in the title
plt.title('Scatter plot using MatplotLib')
# Add in the labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
````

## Scatter plot with groups
Now we will augment the scatter plot to include:
- 3 sets of data
- a legend on the left
- color encoding of the groups
- introduction of the Python `zip()` function that allows us to create a data structure that can be used in our **for** loop

The ``zip()`` function returns a zip object, which is an iterator of tuples where the first item in each passed iterator is paired together, and then the second item in each passed iterator are paired together, etc.


````python
# define the data representing the x,y coordinates of the groups as a tuple of tuples
data = (group1, group2, group3)
# define the colors we want to use for the three groups
colors = ("crimson", "green", "blue")
# add labels to the groups 
groups = ("crimson group", "green group", "blue group")

# Create the scatter plot
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])

# add in each of the data sets in succession
for data, color, group in zip(data, colors, groups):
    x, y = data
    ax.scatter(x, y, alpha=.9, c=color, s=30, label=group)

plt.title('Matplotlib scatter plot with multiple data sets')
# place the legend on the left
plt.legend(loc=2)
# Add in the labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# display the scatter plot
plt.show()
````

 ### Definition of a Histogram  
   
A histogram is a display of statistical information that uses rectangles to show the frequency of data items in successive numerical intervals of equal size. 

In the most common form of histogram, the **independent** variable is plotted along the horizontal axis and the **dependent** variable is plotted along the vertical axis.

<img src="images/histogram_example1.jpg" width="500">

Matplotlib provides the **pyplot.hist()** function for creating histograms plots. 

The **pyplot.hist()** function parameters to be used with axes object are as follows −

``pyplot.hist(x_values, bins, color, transparency)``

## Data set for a Histogram 

For our first data set:
- we will define a sequence of points (list)
- separate the data into 5 bins (the number of intervals to divide the data into)

For our second data set:
- define x as a distribution with mean 100 and sd of 15
- separate the data into 20 bins

````python
import numpy as np

# for the first data set define x and the number of bins
x = [21,22,23,4,5,6,77,8,9,10,31,32,33,34,35,36,37,18,49,50,100]
num_bins = 5

# for the second data set define x as a distribution with mean 100 and sd of 15
mu = 100 # mean of distribution
sigma = 15 # standard deviation of distribution
x_2 = mu + sigma * np.random.randn(10000)
num_bins_2 = 20
````

## Creating the Histogram 

For the histogram, we will use the ``hist`` function in the **pyplot** library.

This simple example includes:
- A call to the hist() function to create the histogram
- The addition of the graphical labels

````python
import matplotlib.pyplot as plt

# use the "hist" function to create the histogram in the default Figure and Axes
plt.hist(x, num_bins, facecolor='blue', alpha=0.5)
# Add in the title
plt.title('Simple Histogram using MatplotLib')
# Add in the labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
# display the graph
plt.show()
````

## Histogram with a fit line

In this example, we use the second data set and create a histogram representing a normal distribution. In addition, we will add a fit line (the data spread across our data range).

Here we see:
- return values from the hist() function
- the plotting of the normally distributed data set
- the addition of a fit line
- the addition of a legend
- our normal labels

````python
import scipy.stats as sp
import matplotlib.lines as mlines

# the histogram of the data - we will use the return values assigned later
n, bins, patches = plt.hist(x_2, num_bins_2, density=1, facecolor='blue', alpha=0.5)

# add a 'best fit' line
y = sp.norm.pdf(bins, mu, sigma)
plt.plot(bins, y, 'r--')

# add in our labels and title
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.title(r'Histogram of IQ: $\mu=100$, $\sigma=15$')

# create a legend
red_line = mlines.Line2D([], [], color='red', marker="_",
                          markersize=15, label='fit line')
plt.legend(handles=[red_line])

# Tweak spacing to prevent clipping of ylabel
plt.subplots_adjust(left=0.15)
# display the plot
plt.show()
````

### Setting Limits
Matplotlib automatically arrives at the minimum and maximum values of variables to be displayed along x, y (and z axis in case of 3D plot) axes of a plot. However, it is possible to set the limits explicitly by using `set_xlim()` and `set_ylim()` functions.


````python
#import numpy
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
# use arrange to generate our data
x = np.arange(1,10)
#plot x and exp(x) as y, color green
a1.plot(x, np.exp(x), 'g')
# don't set any limits
a1.set_title('auto limits')
plt.show()
````

Same plot now, but with the limits set:

````python
#import numpy
import numpy as np
fig = plt.figure()
a1 = fig.add_axes([0,0,1,1])
# use arrange to generate our data
x = np.arange(1,10)
#plot x and exp(x) as y, color red
a1.plot(x, np.exp(x), 'r')
a1.set_title('explicitly set limits')
# set the limits to 10,000 and 10
a1.set_ylim(0,10000)
a1.set_xlim(0,10)
plt.show()
````

### Grids
The `grid()` function of axes object sets visibility of grid inside the figure to on or off. You can also display major / minor (or both) ticks of the grid. Additionally color, linestyle and linewidth properties can be set in the `grid()` function.

````python
# create the figure and subplots - there will be three of them
fig, axes = plt.subplots(1,3, figsize = (12,4))
# create a data set
x = np.arange(1,11)
# plot a line in subplot 1 with the grid on
axes[0].plot(x, x**3, 'g',lw=2)
axes[0].grid(True)
axes[0].set_title('default grid')
# plot a line in subplot 2 with the custom grid on
axes[1].plot(x, np.exp(x), 'r')
axes[1].grid(color='b', ls = '-.', lw = 0.25)
axes[1].set_title('custom grid')
# plot a line in subplot 3 with NO grid
axes[2].plot(x,x)
axes[2].set_title('no grid')
# use side by side layout
fig.tight_layout()
plt.show()
````

 ### Definition of the Line of Best fit
   
The line of best fit (or trendline) is an educated guess about where a linear equation might fall in a set of data plotted on a scatter plot. It helps us to see if there is a relationship or correlation between the two factors being studied. This trendline helps us to predict future events relating to the data being studied.

## Simple line fit to a scatter plot using Least Squares method

The "least squares" method is a form of mathematical regression analysis used to determine the line of best fit for a set of data, providing a visual demonstration of the relationship between the data points. Each point of data represents the relationship between a known independent variable and an unknown dependent variable.

<img src="images/Least_Squares_Regression.png" width="500">

In the following example we will:
- Generate random points for x and y
- Calcualte the best fit line for "N" (500) points using the y=mx+b equation for the line
    - Step 1: For each (x,y) point calculate x2 and xy
    - Step 2: Sum all x, y, x2 and xy, which gives us Σx, Σy, Σx2 and Σxy (Σ means "sum up")
    - Step 3: Calculate Slope m:
        - m =  N Σ(xy) − Σx ΣyN Σ(x2) − (Σx)2 
        - (N is the number of points.)
    - Step 4: Calculate Intercept b:
        - b =  Σy − m ΣxN 
    - Step 5: Assemble the equation of a line
        - y = mx + b


````python
# include numpy for the rand() function
import numpy as np
# Create data
N = 500 # we want 500 points
x1 = np.random.rand(N)
y1 = np.random.rand(N)
# we will use the slope as our "line of best fit" using the least squares method: y=mx+b
# first calculate a denominator
denom1 = x1.dot(x1) - x1.mean() * x1.sum()
# now calulate m and b
m = (x1.dot(y1)-y1.mean()* x1.sum())/denom1
b = (y1.mean() * x1.dot(x1) - x1.mean() * x1.dot(y1))/denom1
# now calculat the best fit y values
ybf1 = m*x1 + b
````

## Displaying the plot
Now that we have the equation of the "best fit" line we can use **matplotlib** to display the scatter plot and the best fit line.


````python
# import Pyplot
import matplotlib.pyplot as plt
# define the colors and shape (area) parameters
colors = (0,0,0)
area = np.pi*3

# Creat the scatter plot by using the scatter function
plt.scatter(x1, y1, s=area, c=colors, alpha=0.5)
# Add in the title
plt.title('Scatter plot using MatplotLib')
# Add in the line of best fit
plt.plot(x1, ybf1, 'b')
# Add in the labels
plt.xlabel('x-axis')
plt.ylabel('y-axis')
plt.show()
````

## Line fit to a generated data set using sklearn and LinearRegression()
As a slightly different example of generating a "best fit" line we will use a linear regression technique from the Python Library **sklearn**. The module is ``LinearRegression``.

In addition, we will be using the sklearn dataset module, ``make_blobs`` to generate isotropic Gaussian blobs for clustering. This will give us a nice set of scatter plot data to use for our "fit lines".


````python
from sklearn.datasets import make_blobs
from sklearn.linear_model import LinearRegression
from matplotlib import pyplot
from pandas import DataFrame

# generate 2d classification dataset
X1, y1 = make_blobs(n_samples=300, centers=3, n_features=2)
# define the scatter plot, dots colored by class value
df = DataFrame(dict(x=X1[:,0], y=X1[:,1], label=y1))
# creat the colors to use
colors = {0:'red', 1:'blue', 2:'green'}
# define our figure and axes
fig, ax = pyplot.subplots()
# group the "like" values in the dataframe (label)
grouped = df.groupby('label')
for key, group in grouped:
    group.plot(ax=ax, kind='scatter', x='x', y='y', label=key, color=colors[key])
# calculate the line fit
X = df.iloc[:, 0].values.reshape(-1, 1)  # values converts it into a numpy array
Y = df.iloc[:, 1].values.reshape(-1, 1)  # -1 means that calculate the dimension of rows, but have 1 column
# create object for the class
linear_regressor = LinearRegression()  
# perform linear regression
linear_regressor.fit(X, Y)  
#call the linear regression function to determine the Y values for best fit
Y_pred = linear_regressor.predict(X)  # make predictions
# add in the best fit line
pyplot.plot(X, Y_pred, color='orange')
# Add in the title
pyplot.title('Best fit using LinearRegression()')
# Add in the labels
pyplot.xlabel('x-axis')
pyplot.ylabel('y-axis')
#show the plot
pyplot.show()
````

A **box plot** which is also known as a whisker plot displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

 ### Definition of a Pair Plot 
   
A “pairs plot” is also known as a scatter matrix. A scatter matrix is a collection of scatterplots organized into a grid (or matrix). Each scatterplot shows the relationship between a pair of variables. Plots are just elaborations on this, showing all variables paired with all the other variables

<img src="images/seaborn-pairplot-1.png" width="300">


## Creating a Pair plot with Matplotlib

While **matplotlib** supports the scatter plot, it does not have a core function that represents a scatter plot matrix (a grid of scatter plots). Several more modern Python visualization libraries have an object that supports this. In the following example, we will show the creation of a function that will create a scatter plot matrix using native matplotlib funtions.


````python
def scatterplot_matrix(data, names, **kwargs):
    """Plots a scatterplot matrix of subplots.  Each row of "data" is plotted
    against other rows, resulting in a nrows by nrows grid of subplots with the
    diagonal subplots labeled with "names".  Additional keyword arguments are
    passed on to matplotlib's "plot" command. Returns the matplotlib figure
    object containg the subplot grid."""
    numvars, numdata = data.shape
    fig, axes = plt.subplots(nrows=numvars, ncols=numvars, figsize=(8,8))
    fig.subplots_adjust(hspace=0.05, wspace=0.05)

    for ax in axes.flat:
        # Hide all ticks and labels
        ax.xaxis.set_visible(False)
        ax.yaxis.set_visible(False)

        # Set up ticks only on one side for the "edge" subplots...
        if ax.is_first_col():
            ax.yaxis.set_ticks_position('left')
        if ax.is_last_col():
            ax.yaxis.set_ticks_position('right')
        if ax.is_first_row():
            ax.xaxis.set_ticks_position('top')
        if ax.is_last_row():
            ax.xaxis.set_ticks_position('bottom')

    # Plot the data.
    for i, j in zip(*np.triu_indices_from(axes, k=1)):
        for x, y in [(i,j), (j,i)]:
            axes[x,y].plot(data[x], data[y], **kwargs)

    # Label the diagonal subplots...
    for i, label in enumerate(names):
        axes[i,i].annotate(label, (0.5, 0.5), xycoords='axes fraction',
                ha='center', va='center')

    # Turn on the proper x or y axes ticks.
    for i, j in zip(range(numvars), itertools.cycle((-1, 0))):
        axes[j,i].xaxis.set_visible(True)
        axes[i,j].yaxis.set_visible(True)

    return fig
````

````python
import matplotlib.pyplot as plt
import numpy as np
import itertools # used for efficient iteration in the scatterplot_matrix function
# generate our random data
np.random.seed(1977)
numvars, numdata = 4, 100
data = 10 * np.random.random((numvars, numdata))
# create our display figure by calling our function that will also create 
# each of the subplot scatter plots - mfc= marker_face_color
fig = scatterplot_matrix(data, ['mpg', 'disp', 'drat', 'wt'],
        linestyle='none', marker='o', color='blue', mfc='red')
# Add a title
fig.suptitle('Simple Scatterplot Matrix')
# Show the scatterplot matrix
plt.show()
````

 ### Definition of a Time Series plot
   
A time series plot is a graph where some measure of time is the unit on the x-axis. In fact, we label the x-axis the time-axis. The y-axis is for the variable that is being measured. Data points are plotted and generally connected with straight lines, which allows for the analysis of the graph generated. 

<img src="images/time_series_graph.png" width="400">


## Creating a Time Series plot using Pandas (and Matplotlib)

A Time Series plot in **matplotlib** is, in essence, a special case of a line plot. That is, we are plotting x and y points connected by a line, however, the x coordinates represent ``time``.

Let's look at an example:


````python
import pandas as pd 
from dateutil.parser import parse
# Import Data
df = pd.read_csv('https://github.com/selva86/datasets/raw/master/AirPassengers.csv')
dates = pd.DatetimeIndex([parse(d).strftime('%Y-%m-01') for d in df['date']])
df.set_index(dates, inplace=True)

# convert the dataframe into a Series, there is only one column
ts = df['value']
ts.head()
````

````python
# visualize the time series with a simple line plot using pandas
ts.plot()
````

## Create a forcasting time series plot using a Seasonal ARIMA model
In the next example, we will use an ARIMA model - AutoRegressive Integrated Moving Average. It is a class of model that captures a suite of different standard temporal structures in time series data. 

Due to the seasonality of Air Travel, we will use an ARIMA model that accounts for Seasonality, named SARIMAX.

We will:
- define the model by analyzing the existing data set
- fit the model
- plot the original data along with the fitted data

````python
import matplotlib.pyplot as plt
# from statsmodels.tsa.arima_model import ARIMA
# include the Seasonal Arima Model for forecasting, SARIMAX
from statsmodels.tsa.statespace.sarimax import SARIMAX 

model = SARIMAX(ts, trend='n', order=(0,1,0), seasonal_order=(1,1,1,12))
results_ARIMA = model.fit(disp=-1)  
plt.plot(ts, label='original')
plt.plot(results_ARIMA.fittedvalues, color='red', label='fitted')
plt.title('RSS: %.4f'% sum((results_ARIMA.fittedvalues-ts)**2))
plt.legend()
````

Now that we have the fitted model, we can plot the original data along with a set of "predicted" data based on a time frame of our choosing:
- for actual data is from 1949 to 1961
- we will "predict" the values from May, 1957 to December, 1960 

````python
#convert our data series back into a dataframe
df = pd.DataFrame(ts)
# Add a forecast column to the dataframe that includes our predictions from the model
df['forecast'] = results_ARIMA.predict(start = '1957-05-01', end= '1960-12-01', dynamic= True)
# Plot the Dataframe data 
df[['value', 'forecast']].plot(figsize=(12, 8))
# Add a title
plt.title('Arima model forecast of #passengers compared to actual')
# Show the plot
plt.show()
````

 ## What are 3D plots in Matplotlib
   
Three-dimensional plots are enabled by importing the **mplot3d** toolkit, included with the Matplotlib package. Three-dimensional axes can be created by passing the keyword `projection='3d'` to any of the normal axes creation routines. 

The 3D capabilitiy is provided by the **mplot3d toolkit** which simplifies 3D plotting capabilities in matplotlib by supplying an axes object that can create a 2D projection of a 3D scene. The resulting graph will have the same look and feel as regular 2D plots.


## Creating a scatter plot in 3D

The **mplot3d toolkit** allows us to make any 2D graph in **matplotlib** into a 3D graph.

The inclusion of ``mplot3d`` enables us to add the third dimension to the `sub_plot` and allows the plotly functions to "graph" on the 3rd axis.

````python
# libraries
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D
# include pyplot
import matplotlib.pyplot as plt
# include numpy
import numpy as np
# include pandas
import pandas as pd
 
# create the Dataset as a dataframe
df=pd.DataFrame({'X': range(1,201), 'Y': np.random.randn(200)*15+range(1,201), 'Z': (np.random.randn(200)*15+range(1,201))*2 })
 
# plot
fig = plt.figure(figsize=(8,8))
# use the 3D projection from the mplot3d import
ax = fig.add_subplot(111, projection='3d')
# not the scatterplot function will be a 3 axis graph
ax.scatter(df['X'], df['Y'], df['Z'], c='blue', s=60)
#show the plot
plt.show()
````

## Creating a bar chart in 3D
In this case we will:
- generate a set of data using ``meshgrid``
- calulate a Z or top value
- plot the 3d and 2d charts

````python
import numpy as np
import matplotlib.pyplot as plt
# This import registers the 3D projection, but is otherwise unused.
from mpl_toolkits.mplot3d import Axes3D  # noqa: F401 unused import


# setup the figure and axes
fig = plt.figure(figsize=(10, 8))
ax1 = fig.add_subplot(221, projection='3d')
ax2 = fig.add_subplot(222, projection='3d')
ax3 = fig.add_subplot(223) #2d
ax4 = fig.add_subplot(224) #2d

# fake data
_x = np.arange(4)
_y = np.arange(5)
# create a grid out of two arrays (cartesian indexing)
_xx, _yy = np.meshgrid(_x, _y)
# flatten the matrix into x and y arrays
x, y = _xx.ravel(), _yy.ravel()

#calculate a Z
top = x + y
bottom = np.zeros_like(top)
width = depth = 1

ax1.bar3d(x, y, bottom, width, depth, top, shade=True)
ax1.set_title('Shaded')

ax2.bar3d(x, y, bottom, width, depth, top, shade=False)
ax2.set_title('Not Shaded')

ax3.bar(x, top, color='red')
ax3.set_title('2D x and z')

ax4.bar(y, top, color='green')
ax4.set_title('2D y and z')


plt.show()
````

 ### Introduction to savefig()
   
**Matplotlib** plots can be saved as image files using the **plt.savefig()** function.

The **plt.savefig()** function needs to be called right above the plt.show() line. All the features of the plot must be specified before the plot is saved as an image file. If the figure is saved after the plt.show() command; the figure will not be saved until the plot window is closed. Calling plt.savefig() after calling plt.show() can be problematic when building plots in a Jupyter notebook with %matplotlib inline enabled.

A standard **savefig()** command is:


``plt.savefig('plot.png', dpi=300, bbox_inches='tight')``

Where ``'plot.png'`` is the name of the saved image file. Matplotlib infers the image file format (.png, .jpg, etc) based on the extension specified in the filename.

The keyword argument ``dpi=`` specifies how many dots per inch (image resolution) are in the saved image. dpi=72 is fine for web images. dpi=300 is better for an image designed to go in a written report or .pdf document.

The keyword argument ``bbox_inches='tight'`` is optional. If the axis labels in the plot are cut off in the saved image, set bbox_inches='tight'.

## Exporting a png file
In this first example, we will use the **savefig()** function to export a figure in **PNG** format.

The image we create is a simple line plot.


````python
import matplotlib.pyplot as plt
# if using a Jupyter notebook, include:
%matplotlib inline

x = [0, 2, 4, 6]
y = [1, 3, 4, 8]


plt.plot(x,y)


plt.xlabel('x values')
plt.ylabel('y values')
plt.title('plotted x and y values')
plt.legend(['line 1'])


# save the figure
plt.savefig('plot.png', dpi=300, bbox_inches='tight')

# as mentioned, the show() need to follow the savefig() in Jupyter
plt.show()
````

## Bar plots


A bar chart or bar graph is a chart or graph that presents categorical data with rectangular bars with heights or lengths proportional to the values that they represent. The bars can be plotted vertically or horizontally.

A bar graph shows comparisons among discrete categories. One axis of the chart shows the specific categories being compared, and the other axis represents a measured value.

For example, we can use bar graphs to compare data between different groups or to track changes over time (similar to a line chart). Multi-series bar graphs can display multiple groups in the same bar in a stacked visual showing the composition of the bar chart along the Y-axis.

Matplotlib provides the **bar()** function for creating bar plots. 

You can use the **bar()** function parameters below with the axes object:

``bar(x, height, width, bottom, align)``


Observe the example below, we have defined the lists `students` and `langs`. We can use a bar plot to visualize which programming languages are more or less popular.

## Histograms
A histogram is a display of statistical information that uses rectangles to show the frequency of data items in successive numerical intervals of equal size. 

Arguably, histograms and bar charts look somewhat familiar. However, it is important that these two are not confused. Histograms are used to show distributions of variables while bar charts are used to compare variables. Histograms plot quantitative data with ranges of the data grouped into bins or intervals while bar charts plot categorical data.

Matplotlib provides the **pyplot.hist()** function for creating histograms plots. 


Observe the example below:

![](./assets/hist.png)

Notice that the horizontal axis of Figure 1 consists of binned times: the first bin includes visits from 0 up to and including ten minutes, the second bin from 10 up to and including 20 minutes, and so on. The vertical axis shows percentages. The area of each bar gives the percentage of all visitors who spent the amount of time shown in the corresponding bin. The sum of all areas equals 100%.

## Box plots

A box plot is a plot that displays a summary of a set of data containing the minimum, first quartile, median, third quartile, and maximum of the data. In a box plot, we draw a box from the first quartile to the third quartile. A vertical line goes through the box at the median. The whiskers go from each quartile to the minimum or maximum.

Usually, box plot are used to display different distributions of quantitative values in a field.

We will use the `boxplot()` function to create a boxplot with Matplotlib. Parameters in the `boxplot()` function determine orientation, colors, labels for the figure.
